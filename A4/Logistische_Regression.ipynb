### Zielvariable: Hat der Schüler bestanden?

In dieser Zelle wird die Zielvariable `bestanden` eingeführt. Ein Schüler gilt als **bestanden**, wenn die Abschlussnote `G3` größer als 10 ist. Daraus wird eine binäre Klassifikationsvariable abgeleitet (1 = bestanden, 0 = nicht bestanden). Anschließend wird die Verteilung dieser Zielklasse ausgegeben.

Diese Zielvariable bildet die Grundlage für die Klassifikationsaufgabe im weiteren Verlauf.

**Hinweis:** Falls `G3` fehlende Werte enthält, wäre eine Vorabprüfung sinnvoll.

```python
# Ziel: Hat der Schüler bestanden? Wir definieren: bestanden = G3 > 10
df['bestanden'] = (df['G3'] > 10).astype(int)
df['bestanden'].value_counts()
```

### One-Hot-Encoding: Umwandlung kategorialer Merkmale

In dieser Zelle werden alle **nicht-numerischen Spalten** des Datensatzes in numerische Form überführt, indem **One-Hot-Encoding** angewendet wird. Dabei wird jede kategoriale Variable in binäre Spalten umgewandelt. Mit `drop_first=True` wird dabei eine Dummy-Falle vermieden, indem jeweils eine Kategorie pro Merkmal weggelassen wird.

Die Zielvariable `y` wird separat gespeichert, und `X` enthält die daraus vorbereiteten Eingabe-Features. Die ursprünglichen Noten `G1`, `G2`, `G3` sowie die Zielvariable `bestanden` werden aus `X` entfernt, da sie entweder Ziel oder stark korreliert mit dem Ziel sind.

```python
# Nicht-numerische Spalten umwandeln (One-Hot-Encoding)
df_encoded = pd.get_dummies(df.drop(columns=['G1', 'G2', 'G3','bestanden']), drop_first=True)
X = df_encoded

y = df['bestanden']
```

### Aufteilung in Trainings- und Testdaten

Hier wird der vorbereitete Datensatz in **Trainings- und Testdaten** aufgeteilt. Mit `train_test_split` aus `sklearn.model_selection` werden:

- 70 % der Daten für das Training (`X_train`, `y_train`)
- 30 % für das Testen (`X_test`, `y_test`)

verwendet. Der Parameter `random_state=42` stellt sicher, dass die Aufteilung reproduzierbar ist.

Diese Trennung ist essenziell, um ein Modell unabhängig zu trainieren und anschließend seine Generalisierungsfähigkeit zu bewerten.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

### Modelltraining mit logistischer Regression

Hier wird ein **logistisches Regressionsmodell** aus `sklearn.linear_model` erstellt und mit den Trainingsdaten trainiert:

- `LogisticRegression(max_iter=1000)` legt fest, dass die Optimierung bis zu 1000 Iterationen erlaubt ist (Standard ist 100, was bei vielen Features evtl. nicht ausreicht).
- `model.fit(X_train, y_train)` führt das Training durch.

Die logistische Regression ist ein einfaches, gut interpretierbares Modell zur **binären Klassifikation** – in diesem Fall: *Hat der Schüler bestanden (1) oder nicht (0)?*

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

### Modellbewertung: Konfusionsmatrix und Klassifikationsbericht

Nach dem Training wird das Modell auf den Testdaten evaluiert:

- `model.predict(X_test)` erzeugt Vorhersagen für die Testmenge.
- `confusion_matrix(y_test, y_pred)` zeigt, wie viele Schüler korrekt/nicht korrekt als bestanden/nicht bestanden klassifiziert wurden.
- `classification_report(y_test, y_pred)` liefert präzise Kennzahlen wie **Precision**, **Recall** und **F1-Score** für beide Klassen.

Diese Auswertung hilft zu beurteilen, wie gut das Modell im Klassifikationsproblem "Bestanden oder nicht?" abschneidet.

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

### Wichtigste Merkmale: Einfluss auf das Bestehen

Mit `model.coef_` lassen sich die **gelernten Gewichtungen** der logistischen Regression auslesen. Diese geben an, wie stark ein Merkmal (positiv oder negativ) die Wahrscheinlichkeit beeinflusst, dass ein Schüler besteht.

- Die `Series` wird nach den höchsten Koeffizienten sortiert.
- Die Top 10 Merkmale mit **positivem Einfluss** auf das Bestehen werden angezeigt.

Dies ermöglicht eine erste Interpretation, welche Faktoren laut Modell besonders relevant sind.

```python
coeffs = pd.Series(model.coef_[0], index=X.columns)
coeffs.sort_values(ascending=False).head(10)
```

### ROC-Kurve und AUC: Modellgüte visualisieren

Die **ROC-Kurve** (Receiver Operating Characteristic) zeigt die Leistung des Klassifikators bei verschiedenen Schwellenwerten. Je näher die Kurve an der linken oberen Ecke verläuft, desto besser das Modell.

- Die **True Positive Rate (TPR)** wird gegen die **False Positive Rate (FPR)** aufgetragen.
- Der **AUC-Wert** (Area Under Curve) quantifiziert die Modellgüte (1 = perfekt, 0.5 = Zufall).
- `predict_proba()` liefert die geschätzte Wahrscheinlichkeit für Klasse 1 („bestanden“).

Eine gute Klassifikation zeigt sich in einem hohen AUC-Wert (> 0.8 ist i.d.R. stark).

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score

probs = model.predict_proba(X_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, probs)
auc = roc_auc_score(y_test, probs)

plt.plot(fpr, tpr, label=f'AUC = {auc:.2f}')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-Kurve')
plt.legend()
plt.grid(True)
plt.show()
```

### Multiklassen-Klassifikation 

Die logistische Regression liefert uns Ergebnisse für binäre Zielvariablen.
Oftmals wollen wir aber mehr als 2 Klassen unterscheiden.

Eine Möglichkeit, um Multiklassen-Klassifikation mit logistischen Regression umzusetzen, ist die sogenannte *One-vs-all Klassifikation*.
Dabei werden für `n` Klassen `n` separate, binäre Klassifikationsprobleme definiert, bei denen jeweils nur die betrachtete Klasse den Zielwert `1` zugeteilt bekommt, und für alle anderen Klassen der Zielwert `0` angenommen wird.

Scikit-Learn unterstüzt Multiklassen-Klassifikation in der Klasse `LogisticRegression` über den Parameter `multi_class`.
Setzt man : `multi_class="ovr"` führt die Funktion `fit` je eine logistische Regression für jedes Label nach dem *one-vs-all* (oder auch *one-vs-rest*, ovr) Prinzip aus.

In der folgenden Code-Zelle erzeugen wir 3 Punktwolken.
Alle Punkte einer "Wolke" sollen zu einer bestimmten Klasse gehören.

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from pandas import DataFrame
# generate 2d classification dataset
X, y = make_blobs(n_samples=100, centers=3, n_features=2, random_state=10, cluster_std=2.5)
# scatter plot, dots colored by class value
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
colors = {0:'red', 1:'blue', 2:'green'}
markers = {0:'o', 1:'x', 2:'^'}
fig, ax = plt.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, marker=markers[key], color=colors[key])
    
plt.legend(loc='upper right', prop={'size': 12})
plt.savefig("LogistischeRegression20.png",transparent=True, dpi=300)
plt.show()
```

Nun wenden wir ein logistisches Regressionsmodell auf die Datenbasis an.

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=0)

model = LogisticRegression(solver='lbfgs',multi_class="ovr")

model.fit(X_train,y_train)

model.intercept_ , model.coef_
```

Der folgende Graph zeigt die Entscheidungsgrenzen für das Klassifikationsmodell.
Alle Punkte innerhalb eines Bereiches werden der jeweiligen Klasse zugeordnet.
